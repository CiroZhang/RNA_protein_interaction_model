```python
from genechat_model import GeneChatModel
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
model = GeneChatModel(freeze_encoders=True).to(device)

dna_seq = "ATGGCTCAGGTTAGCTGATGCTAGCAGTTGCGTACG"
prompt = "Describe the gene's function."

# Dummy input for decoder during training
dummy_input = model.text_tokenizer("start", return_tensors="pt").input_ids.to(device)

# Forward pass (for training)
logits = model(dna_seq, prompt, dummy_input, device)

# Or text generation (for demo)
text = model.generate(dna_seq, prompt, device=device)
print("Generated text:", text)
```
